```python
import cv2
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# computer vision
img = cv2.imread('picture 1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)

t = np.array([[-1,0,-1],
             [0,5,0],
             [-1,0,-1]])
result = cv2.filter2D(img,-1,t)
result = np.hstack((img,result))
plt.imshow(result)
result = cv2.cvtColor(result,cv2.COLOR_RGB2BGR)
cv2.imwrite('out1.jpg',result)
```

```python
t = np.array([[-2,-1,0],
             [-1,1,1],
             [0,2,1]])
result = cv2.filter2D(img,-1,t)
result = np.hstack((img,result))
plt.imshow(result)
result = cv2.cvtColor(result,cv2.COLOR_RGB2BGR)
cv2.imwrite('out2.jpg',result)
```
